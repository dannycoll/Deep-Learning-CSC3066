```python
def encryptRailFence(text, key):
	n = len(text)
	m = key
	rail = [['\n' for i in range(n)]
				for j in range(m)]
	dir_down = False
	row, col = 0, 0
	for i in range(n):
		if (row == 0) or (row == m - 1):
			dir_down = not dir_down
		rail[row][col] = text[i]
		col += 1
		if dir_down:
			row += 1
		else:
			row -= 1
	result = []
	for i in range(m):
		for j in range(n):
			if rail[i][j] != '\n':
				result.append(rail[i][j])
	return("" . join(result))
	
def decryptRailFence(cipher, key):
	n = len(cipher)
	m = key
	rail = [['\n' for i in range(n)]
				for j in range(m)]
	dir_down = None
	row, col = 0, 0
	for i in range(n):
		if row == 0:
			dir_down = True
		if row == m - 1:
			dir_down = False
		rail[row][col] = '*'
		col += 1
		if dir_down:
			row += 1
		else:
			row -= 1
	index = 0
	for i in range(m):
		for j in range(n):
			if ((rail[i][j] == '*') and (index < n)):
				rail[i][j] = cipher[index]
				index += 1
	result = []
	row, col = 0, 0
	for i in range(n):
		if row == 0:
			dir_down = True
		if row == m-1:
			dir_down = False
		if (rail[row][col] != '*'):
			result.append(rail[row][col])
			col += 1
		if dir_down:
			row += 1
		else:
			row -= 1
	return("".join(result))

def encrypt(text,s):
    result = ""
 
    # traverse text
    for i in range(len(text)):
        char = text[i]
 
        # Encrypt uppercase characters
        if (char.isupper()):
            result += chr((ord(char) + s-65) % 26 + 65)
 
        # Encrypt lowercase characters
        else:
            result += chr((ord(char) + s - 97) % 26 + 97)
 
    return result

if __name__ == "__main__":
	text = 'xf ie ttorqw xnoxg seneos hhwu u hen irxat ttdfrsfe omxexomis m mhtfho ieo siyoxsx ag etaneghi stfa se t t wghxceto iefeih sneimthrino txrin peoreeon'
	no_spaces = 'xreiotstxhhxommnffsdsseeeyxqwogtsfaaietierehrmoruxwixentgghepoiatiiueexonxciixhteeffsfewrhsnnertoxtsnhteoosomttogmenithnhoia'
	for j in range(20):
		print(decryptRailFence(no_spaces,j+2))
		print(j+2)

```

```python
def heap(a):
  n = len(a)
  c = [0]*n
  A = list(a)
  yield A
  i = 1
  while i < n:
    if c[i] < i:
      if i % 2 == 0:
        A[0],A[i] = A[i],A[0]
      else:
        A[c[i]],A[i]=A[i],A[c[i]]
      yield A
      c[i] += 1
      i = 1
    else:
      c[i] = 0
      i += 1
```

```python
def table2cycle(perm):
    to_do=perm[:]
    cycles=[]      
    while len(to_do)>0: 
        cycle=[]
        image=to_do[0]  
        while not (image in cycle):
            cycle.append(image)
            to_do.remove(image)
            image=perm[image]  
        cycles.append(cycle)  
    return cycles             
```
